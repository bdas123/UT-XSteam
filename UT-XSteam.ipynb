# Using XSteam

XSteam takes 2 or 3 arguments. The first argument must always be the steam table function you want to use. <br>
The other arguments are the inputs to that function. <br>
example: XSteam('h_pt',1,20)  Returns the enthalpy of water at 1 bar and 20 degC <br>
example: XSteam('TSat_p',1)  Returns the saturation temperature of water at 1 bar. <br>
For a list of valid Steam Table functions se bellow or the XSteam macros for MS excel. <br>

```python
def to_systeme_international_units_pressure(input_temperature_bar):
    temperature_mpa = input_temperature_bar/10
    return temperature_mpa 

def from_system_international_units_pressure(input_temperature_mpa):
    temperature_bar = input_temperature_mpa * 10
    return temperature_bar

def to_systeme_international_units_temperature(input_temperature_celsius):
    temperature_kelvin = input_temperature_celsius + 273.15
    return temperature_kelvin
    
def from_systeme_international_units_temperature(input_temperature_kelvin):
    output_temperature_celsius = input_temperature_kelvin - 273.15
    return output_temperature_celsius
```

```python
def p4_T(T):
    # Release on the IAPWS Industrial formulation 1997 for the Thermodynamic Properties of Water and Steam, September 1997
    # Section 8.1 The Saturation-Pressure equation
    # eq 30, Page 33
    teta = T - 0.23855557567849 / (T - 650.17534844798);
    a = (teta**2) + 1167.0521452767 * teta - 724213.16703206;
    B = -17.073846940092 * (teta**2) + 12020.82470247 * teta - 3232555.0322333;
    C = 14.91510861353 * (teta**2) - 4823.2657361591 * teta + 405113.40542057;
    p4_T = (2 * C / (-B + (B**2 - 4 * a * C)**0.5))**4;
    return p4_T
```

```python
def T4_p(p):
    # Release on the IAPWS Industrial formulation 1997 for the Thermodynamic Properties of Water and Steam, September 1997
    # Section 8.2 The Saturation-Temperature equation
    # eq 31, Page 34
    beta = (p ** 0.25);
    e = beta ** 2 - 17.073846940092 * beta + 14.91510861353;
    f = 1167.0521452767 * (beta ** 2) + 12020.82470247 * beta - 4823.2657361591;
    G = -724213.16703206 * (beta ** 2) - 3232555.0322333 * beta + 405113.40542057;
    D = 2 * G / (-f - (f**2 - 4 * e * G)**0.5);
    T4_p = (650.17534844798 + D - ((650.17534844798 + D) ** 2 - 4 * (-0.23855557567849 + 650.17534844798 * D)) ** 0.5) / 2
    return T4_p
```

```python
def w5_pT(p, T):
    # Release on the IAPWS Industrial formulation 1997 for the Thermodynamic Properties of Water and Steam, September 1997
    # Basic equation for Region 5
    # eq 32,33, Page 36, Tables 37-41
    Ji0 = np.array([0, 1, -3, -2, -1, 2])
    ni0 = np.array([-13.179983674201, 6.8540841634434, -0.024805148933466, 0.36901534980333, -3.1161318213925, -0.32961626538917])
    Iir = np.array([1, 1, 1, 2, 3])
    Jir = np.array([0, 1, 3, 9, 3])
    nir = np.array([-1.2563183589592e-04, 2.1774678714571e-03, -0.004594282089991, -3.9724828359569e-06, 1.2919228289784e-07])
    R = 0.461526; #kJ/(kg K)
    tau = 1000 / T;
    Pi = p;
    gamma0_tautau = 0;
    gamma0_tautau += np.sum(ni0 * (Ji0 - 1) * Ji0 * (tau**(Ji0 - 2)))
    gammar_pi = 0;
    gammar_pitau = 0;
    gammar_pipi = 0;
    gammar_tautau = 0;
    gammar_pi += np.sum(nir * Iir * (Pi**(Iir - 1)) * (tau**Jir))
    gammar_pitau += np.sum(nir * Iir * (Pi**(Iir - 1)) * Jir * (tau**(Jir - 1)))
    gammar_pipi += np.sum(nir * Iir * (Iir - 1) * Pi**(Iir - 2) * (tau**Jir));
    gammar_tautau += np.sum(nir * (Pi**Iir) * Jir * ((Jir - 1)) * (tau**(Jir - 2)));
    w5_pT = (1000 * R * T * (1 + 2 * Pi * gammar_pi + (Pi**2) * (gammar_pi**2)) / ((1 - (Pi**2) * gammar_pipi) + ((1 + Pi * gammar_pi - tau * Pi * gammar_pitau)**2) / (tau ** 2 * (gamma0_tautau + gammar_tautau))))**0.5;
    return w5_pT
    
    
```

```python
def T5_ph(p, h):
    # Solve with half interval method
    Low_Bound = 1073.15;
    High_Bound = 2273.15;
    hs=h-1;
    while np.abs(h - hs) > 0.00001:
        Ts = (Low_Bound + High_Bound) / 2;
        hs = h5_pT(p, Ts);
        if hs > h: 
            High_Bound = Ts;
        else:
            Low_Bound = Ts;
    T5_ph = Ts
    return T5_ph
```

```python
def T5_ps(p, s):
    #Solve with half interval method
    Low_Bound = 1073.15;
    High_Bound = 2273.15;
    ss=s-1;
    while abs(s - ss) > 0.00001:
        Ts = (Low_Bound + High_Bound) / 2;
        ss = s5_pT(p, Ts);
        if ss > s: 
            High_Bound = Ts;
        else:
            Low_Bound = Ts;
    T5_ps = Ts
    return T5_ps
```

```python
#3 Region Selection
#3.1 Regions as a function of pT
def region_pT(p, T):
    if T > 1073.15 and p < 10 and T < 2273.15 and p > 0.000611: 
        region_pT = 5;
    elif T <= 1073.15 and T > 273.15 and p <= 100 and p > 0.000611: 
        if T > 623.15: 
            if p > B23p_T(T): 
                region_pT = 3;
                if T < 647.096: 
                    ps = p4_T(T);
                    if abs(p - ps) < 0.00001:
                        region_pT = 4;
            else:
                region_pT = 2;
        else:
            ps = p4_T(T);
            if abs(p - ps) < 0.00001: 
                region_pT = 4;
            elif p > ps:
                region_pT = 1;
            else:
                region_pT = 2;
    else:
        region_pT = 0; #error, Outside valid area
    
    return region_pT
```

```python
# 3.2 Regions as a function of ph
def region_ph(p,h):
    # Check if outside pressure limits
    if p < 0.000611657 or p > 100: 
        region_ph = 0;

    # Check if outside low h.
    elif h < 0.963 * p + 2.2:  #Linear adaption to h1_pt()+2 to speed up calcualations.
        if h < h1_pT(p, 273.15): 
            region_ph = 0;

    elif p < 16.5292:  #Below region 3,Check  region 1,4,2,5
        # Check Region 1
        Ts = T4_p(p)
        hL = 109.6635 * np.log(p) + 40.3481 * p + 734.58; # Approximate function for hL_p
        if np.abs(h - hL) < 100:  # if approximate is not god enough use real function
            hL = h1_pT(p, Ts);
        elif h <= hL: 
            region_ph = 1;
        # Check Region 4
        hV = 45.1768 * np.log(p) - 20.158 * p + 2804.4; #Approximate function for hV_p
        if np.abs(h - hL) < 50:  #if approximate is not good enough use real function
            hV = h2_pT(p, Ts)
        elif h < hV: 
            region_ph = 4;
        # Check upper limit of region 2 Quick Test
        elif h < 4000: 
            region_ph = 2;
        # Check region 2 (Real value)
        h_45 = h2_pT(p, 1073.15);
        if h <= h_45: 
            region_ph = 2;
        #Check region 5
        elif p > 10: 
            region_ph = 0;
        h_5u = h5_pT(p, 2273.15);
        if h < h_5u: 
            region_ph = 5;
        region_ph = 0;
        
    else: #for p>16.5292
        
        #Check if in region1
        
        if h < h1_pT(p, 623.15):
            
            region_ph = 1;
            
        # Check if in region 3 or 4 (Bellow Reg 2)
        
        elif h < h2_pT(p, B23T_p(p)): 
            
            # Region 3 or 4
            
            if p > p3sat_h(h): 
                
                region_ph = 3;
                
            else:
                
                region_ph = 4;
                
        # Check if region 2
        
        if h < h2_pT(p, 1073.15): 
            
            region_ph = 2;
    
    return region_ph
```

```python
# 3.3 Regions as a function of ps
def region_ps(p,s):
    if p < 0.000611657 or p > 100 or s < 0 or s > s5_pT(p, 2273.15): 
        region_ps = 0;
        
    # Check region 5
    if s > s2_pT(p, 1073.15): 
        if p <= 10: 
            region_ps = 5;
        else:
            region_ps = 0;

    # Check region 2
    if p > 16.529: 
        ss = s2_pT(p, B23T_p(p)); # Between 5.047  and 5.261. Use to speed up!
    else:
        ss = s2_pT(p, T4_p(p));
        
    if s > ss: 
        region_ps = 2;

    #Check region 3
    
    ss = s1_pT(p, 623.15);
    
    if p > 16.529  and s > ss: 
        if p > p3sat_s(s): 
            region_ps = 3;
        else:
            region_ps = 4;

    # Check region 4 (Not inside region 3)
    if p < 16.529  and s > s1_pT(p, T4_p(p)): 
        region_ps = 4;

    # Check region 1
    if p > 0.000611657  and s > s1_pT(p, 273.15): 
        region_ps = 1;
    
    return region_ps
```

```python
#3.4 Regions as a function of hs

def region_hs(h,s):
    if s < -0.0001545495919: 
        region_hs = 0;
    # Check linear adaption to p=0.000611. if bellow region 4.
    hMin = (((-0.0415878 - 2500.89262) / (-0.00015455 - 9.155759)) * s);
    if s < 9.155759395  and h < hMin: 
        region_hs = 0;

    # Kolla 1 eller 4. (+liten bit över B13)
    if s >= -0.0001545495919  and s <= 3.77828134: 
        if h < h4_s(s): 
            region_hs = 4;
        elif s < 3.397782955:  #100MPa line is limiting
            TMax = T1_ps(100, s);
            hMax = h1_pT(100, TMax);
            if h < hMax: 
                region_hs = 1;
            else:
                region_hs = 0;
        else: # The point is either in region 4,1,3. Check B23
            hB = hB13_s(s);
            if h < hB: 
                region_hs = 1;
            TMax = T3_ps(100, s);
            vmax = v3_ps(100, s);
            hMax = h3_rhoT(1 / vmax, TMax);
            if h < hMax: 
                region_hs = 3;
            else:
                region_hs = 0;
                
    # Kolla region 2 eller 4. (Övre delen av område b23-> max)
    
    if s >= 5.260578707  and s <= 11.9212156897728: 
        if s > 9.155759395:  # Above region 4
            Tmin = T2_ps(0.000611, s);
            hMin = h2_pT(0.000611, Tmin);
            # function adapted to h(1073.15,s)
            hMax = -0.07554022 * (s ** 4) + 3.341571 * (s ** 3) - 55.42151 * (s ** 2) + 408.515 * s + 3031.338;
            if h > hMin  and h < hMax: 
                region_hs = 2;
            else:
                region_hs = 0;

        hV = h4_s(s)

        if h < hV:   # Region 4. Under region 3.
            region_hs = 4;
            
        if s < 6.04048367171238: 
            TMax = T2_ps(100, s);
            hMax = h2_pT(100, TMax);
            
        else:
            # function adapted to h(1073.15,s)
            hMax = -2.988734 * (s**4) + 121.4015 * (s**3) - 1805.15 * (s**2) + 11720.16 * s - 23998.33;
        if h < hMax:   # Region 2. Över region 4.
            region_hs = 2;
        else:
            region_hs = 0;

    # Kolla region 3 eller 4. Under kritiska punkten.
    
    if s >= 3.77828134  and s <= 4.41202148223476: 
        hL = h4_s(s);
        if h < hL: 
            region_hs = 4;

        TMax = T3_ps(100, s);
        vmax = v3_ps(100, s);
        hMax = h3_rhoT(1 / vmax, TMax);
        if h < hMax: 
            region_hs = 3;

        else:
            region_hs = 0;

    # Kolla region 3 eller 4 från kritiska punkten till övre delen av b23
    
    if s >= 4.41202148223476  and s <= 5.260578707: 
        hV = h4_s(s);
        if h < hV: 
            region_hs = 4;

        # Kolla om vi är under b23 giltighetsområde.
        if s <= 5.048096828: 
            TMax = T3_ps(100, s);
            vmax = v3_ps(100, s);
            hMax = h3_rhoT(1 / vmax, TMax);
            if h < hMax: 
                region_hs = 3;

            else:
                region_hs = 0;

        else: # Inom området för B23 i s led.
            if (h > 2812.942061):  #Ovanför B23 i h_led
                if s > 5.09796573397125: 
                    TMax = T2_ps(100, s);
                    hMax = h2_pT(100, TMax);
                    if h < hMax: 
                        region_hs = 2;

                    else:
                        region_hs = 0;

                else:
                    region_hs = 0;
                    
            if (h < 2563.592004):    # Nedanför B23 i h_led men vi har redan kollat ovanför hV2c3b
                region_hs = 3;

            # Vi är inom b23 området i både s och h led.
            Tact = TB23_hs(h, s);
            pact = p2_hs(h, s);
            pBound = B23p_T(Tact);
            if pact > pBound: 
                region_hs = 3;

            else:
                region_hs = 2;
    return region_hs
```

```python
# 3.5 Regions as a function of p and rho
def Region_prho(p,rho):
    v = 1 / rho;
    if p < 0.000611657 or p > 100:
        Region_prho = 0;

    if p < 16.5292:  # Bellow region 3, Check region 1,4,2
        if v < v1_pT(p, 273.15): #Observe that this is not actually min of v. Not valid Water of 4°C is ligther.
            Region_prho = 0;
            
        elif v <= v1_pT(p, T4_p(p)):
            Region_prho = 1;

        elif v < v2_pT(p, T4_p(p)): 
            Region_prho = 4;

        elif v <= v2_pT(p, 1073.15):
            Region_prho = 2;

        elif p > 10: #Above region 5
            Region_prho = 0;

        elif v <= v5_pT(p, 2073.15): 
            Region_prho = 5;

    else: #Check region 1,3,4,3,2 (Above the lowest point of region 3.)
        if v < v1_pT(p, 273.15): #Observe that this is not actually min of v. Not valid Water of 4°C is ligther.
            Region_prho = 0;

        elif v < v1_pT(p, 623.15):
            Region_prho = 1;

        # Check if in region 3 or 4 (Bellow Reg 2)
        elif v < v2_pT(p, B23T_p(p)):
            # Region 3 or 4
            if p > 22.064: #Above region 4
                Region_prho = 3;

            if v < v3_ph(p, h4L_p(p)) | v > v3_ph(p, h4V_p(p)): #Uses iteration!!
                Region_prho = 3;

            else:
                Region_prho = 4;

        # Check if region 2
        if v < v2_pT(p, 1073.15):
            Region_prho = 2;

    return Region_prho;
```

```python
# 4.1 Boundary between region 2 and 3.
def B23p_T(T):
    # Release on the IAPWS Industrial formulation 1997 for the Thermodynamic Properties of Water and Steam
    # 1997
    # Section 4 Auxiliary equation for the Boundary between Regions 2 and 3
    # eq 5, Page 5
    B23p_T = 348.05185628969 - 1.1671859879975 * T + 1.0192970039326e-03 * T ** 2
    
    return B23p_T
```

```python
def B23T_p(p):
    # Release on the IAPWS Industrial formulation 1997 for the Thermodynamic Properties of Water and Steam
    # 1997
    # Section 4 Auxiliary equation for the Boundary between Regions 2 and 3
    # eq 6, Page 6
    B23T_p = 572.54459862746 + ((p - 13.91883977887) / 1.0192970039326e-03) ** 0.5;
    
    return B23T_p
    
```

```python
# 4.2 Region 3. pSat_h  and pSat_s
def p3sat_h(h):
    # Revised Supplementary Release on Backward equations for the functions T(p,h), v(p,h)  and T(p,s), v(p,s) for Region 3 of the IAPWS Industrial formulation 1997 for the Thermodynamic Properties of Water  and Steam
    # 2004
    # Section 4 Boundary equations psat(h)  and psat(s) for the Saturation Lines of Region 3
    # Se pictures Page 17, eq 10, Table 17, Page 18
    Ii = np.array([0, 1, 1, 1, 1, 5, 7, 8, 14, 20, 22, 24, 28, 36])
    Ji = np.array([0, 1, 3, 4, 36, 3, 0, 24, 16, 16, 3, 18, 8, 24])
    ni = np.array([0.600073641753024, -9.36203654849857, 24.6590798594147, -107.014222858224, -91582131580576.8, -8623.32011700662, -23.5837344740032, 2.52304969384128e+17, -3.89718771997719e+18, -3.33775713645296e+22, 35649946963.6328, -1.48547544720641e+26, 3.30611514838798e+18, 8.13641294467829e+37])
    hs = h / 2600;
    ps = 0;
    ps += np.sum(ni * ((hs - 1.02) ** Ii) * ((hs - 0.608) ** Ji))
    
    p3sat_h = ps * 22
    return p3sat_h;
```

```python
def p3sat_s(s):
    Ii = np.array([0, 1, 1, 4, 12, 12, 16, 24, 28, 32])
    Ji = np.array([0, 1, 32, 7, 4, 14, 36, 10, 0, 18])
    ni = np.array([0.639767553612785, -12.9727445396014, -2.24595125848403e+15, 1774667.41801846, 7170793495.71538, -3.78829107169011e+17, -9.55586736431328e+34, 1.87269814676188e+23, 119254746466.473, 1.10649277244882e+36])
    Sigma = s / 5.2;
    Pi = 0;
    Pi += np.sum(ni * ((Sigma - 1.03) ** Ii) * ((Sigma - 0.699)**Ji))
    p3sat_s = Pi * 22
    return p3sat_s
```

```python
# 4.3 Region boundary 1to3  and 3to2 as a functions of s
def hB13_s(s):
    # Supplementary Release on Backward equations ( ) , p h s for Region 3,
    # Chapter 4.5 page 23.
    Ii = np.array([0, 1, 1, 3, 5, 6])
    Ji = np.array([0, -2, 2, -12, -4, -3])
    ni = np.array([0.913965547600543, -4.30944856041991e-05, 60.3235694765419, 1.17518273082168e-18, 0.220000904781292, -69.0815545851641])
    Sigma = s / 3.8;
    eta = 0;
    eta += np.sum(ni * ((Sigma - 0.884) ** Ii) * ((Sigma - 0.864)**Ji))
    hB13_s = eta * 1700
    
    return hB13_s
```

```python
def my_AllRegions_pT(  p,   T):
    h0 = np.array([0.5132047, 0.3205656, 0, 0, -0.7782567, 0.1885447])
    h1 = np.array([0.2151778, 0.7317883, 1.241044, 1.476783, 0, 0])
    h2 = np.array([-0.2818107, -1.070786, -1.263184, 0, 0, 0])
    h3 = np.array([0.1778064, 0.460504, 0.2340379, -0.4924179, 0, 0])
    h4 = np.array([-0.0417661, 0, 0, 0.1600435, 0, 0])
    h5 = np.array([0, -0.01578386, 0, 0, 0, 0])
    h6 = np.array([0, 0, 0, -0.003629481, 0, 0])

    # Calcualte density.
    if region_pT(p,T) == 1:
        rho = 1 / v1_pT(p, T);
    elif region_pT(p,T) == 2:
        rho = 1 / v2_pT(p, T);
    elif region_pT(p,T) == 3:
        hs = h3_pT(p, T);
        rho = 1 / v3_ph(p, hs);
    elif region_pT(p,T) == 4:
        rho = np.nan;
    elif region_pT(p,T) == 5:
        rho = 1 / v5_pT(p, T);
    else:
        my_AllRegions_pT = np.nan;
    

    rhos = rho / 317.763;
    Ts = T / 647.226;
    ps = p / 22.115;

    #Check valid area
    
    if T > 900 + 273.15 | (T > 600 + 273.15  and p > 300) | (T > 150 + 273.15  and p > 350) | p > 500: 
        my_AllRegions_pT = np.nan

    my0 = (Ts**0.5) / (1 + (0.978197 / Ts) + (0.579829 / (Ts**2)) - (0.202354 / (Ts**3)));
    Sum = 0;
    for i in range(0,6):
        Sum = Sum + h0(i+1) * ((1 / Ts - 1)**i) + h1(i+1) * ((1 / Ts - 1) ** i) * ((rhos - 1) ** 1) + h2(i+1) * ((1 / Ts - 1) ** i) * ((rhos - 1) ** 2) + h3(i+1) * ((1 / Ts - 1) ** i) * ((rhos - 1) ** 3) + h4(i+1) * ((1 / Ts - 1) ** i) * ((rhos - 1)**4) + h5(i+1) * ((1 / Ts - 1)**i) * ((rhos - 1)** 5) + h6(i+1) * ((1 / Ts - 1)**i) * ((rhos - 1)**6);
    my1 = np.exp(rhos * Sum);
    mys = my0 * my1;
    my_AllRegions_pT = mys * 0.000055071
    
    return my_AllRegions_pT
```

```python
def my_AllRegions_ph(p,h):
    h0 = np.array([0.5132047, 0.3205656, 0, 0, -0.7782567, 0.1885447])
    h1 = np.array([0.2151778, 0.7317883, 1.241044, 1.476783, 0, 0])
    h2 = np.array([-0.2818107, -1.070786, -1.263184, 0, 0, 0])
    h3 = np.array([0.1778064, 0.460504, 0.2340379, -0.4924179, 0, 0])
    h4 = np.array([-0.0417661, 0, 0, 0.1600435, 0, 0])
    h5 = np.array([0, -0.01578386, 0, 0, 0, 0])
    h6 = np.array([0, 0, 0, -0.003629481, 0, 0])

    # Calcualte density.
    if region_ph(p, h) == 1:
        Ts = T1_ph(p, h);
        T = Ts;
        rho = 1 / v1_pT(p, Ts);
    elif region_ph(p, h) == 2:
        Ts = T2_ph(p, h);
        T = Ts;
        rho = 1 / v2_pT(p, Ts);
    elif region_ph(p, h) == 3:
        rho = 1 / v3_ph(p, h);
        T = T3_ph(p, h);
    elif region_ph(p, h) == 4:
        xs = x4_ph(p, h);
        if p < 16.529: 
            v4v = v2_pT(p, T4_p(p));
            v4L = v1_pT(p, T4_p(p));
        else:
            v4v = v3_ph(p, h4V_p(p));
            v4L = v3_ph(p, h4L_p(p));
        rho = 1 / (xs * v4v + (1 - xs) * v4L);
        T = T4_p(p);
    elif region_ph(p, h) == 5:
        Ts = T5_ph(p, h);
        T = Ts;
        rho = 1 / v5_pT(p, Ts);
    else:
        my_AllRegions_ph = np.nan;
        
    rhos = rho / 317.763;
    Ts = T / 647.226;
    ps = p / 22.115;
    # Check valid area
    if T > 900 + 273.15 | (T > 600 + 273.15  and p > 300) | (T > 150 + 273.15  and p > 350) | p > 500: 
        my_AllRegions_ph = np.nan;

    my0 = Ts ** 0.5 / (1 + 0.978197 / Ts + 0.579829 / (Ts ** 2) - 0.202354 / (Ts ** 3));

    Sum = 0;
    for i in range(0,6):
        Sum += h0[i+1] * (1 / Ts - 1) ** i + h1[i+1] * (1 / Ts - 1) ** i * (rhos - 1) ** 1 + h2[i+1] * (1 / Ts - 1) ** i * (rhos - 1) ** 2 + h3[i+1] * (1 / Ts - 1) ** i * (rhos - 1) ** 3 + h4[i+1] * (1 / Ts - 1) ** i * (rhos - 1) ** 4 + h5[i+1] * (1 / Ts - 1) ** i * (rhos - 1) ** 5 + h6[i+1] * (1 / Ts - 1) ** i * (rhos - 1) ** 6;
    my1 = np.exp(rhos * Sum);
    mys = my0 * my1;
    my_AllRegions_ph = mys * 0.000055071
    return myAllRegions_ph
```

##  5.2 Thermal Conductivity (IAPWS formulation 1985)

```python
def tc_ptrho(p,T,rho):
    # Revised release on the IAPWS formulation 1985 for the Thermal Conductivity of ordinary water
    # IAPWS September 1998
    # Page 8
#    if T < 0 | p < 0.000611657 | T > 800 | p > 400 | not((p <= 100  and T <= 100 + 273.15) | (p <= 150  and T <= 400 + 273.15) | (p <= 200  and T <= 250 + 273.15) | (p <= 400  and T <= 125 + 273.15): 
#        tc_ptrho = np.nan; #'Out of valid region'
        
    T = T / 647.26;
    rho = rho / 317.7;
    tc0 = (T ** 0.5) * (0.0102811 + 0.0299621 * T + 0.0156146 * (T ** 2) - 0.00422464 * (T**3));
    tc1 = -0.39707 + 0.400302 * rho + 1.06 * np.exp(-0.171587 * (rho + 2.39219) ** 2);
    dT = np.abs(T - 1) + 0.00308976;
    Q = 2 + (0.0822994 / dT) ** (3 / 5);
    if T >= 1: 
        s = 1 / dT
    else:
        s = 10.0932 / (dT ** (3 / 5));
    tc2 = (0.0701309 / T ** 10 + 0.011852) * rho ** (9 / 5) * np.exp(0.642857 * (1 - rho ** (14 / 5))) + 0.00169937 * s * rho ** Q * np.exp((Q / (1 + Q)) * (1 - rho ** (1 + Q))) - 1.02 * np.exp(-4.11717 * T ** (3 / 2) - 6.17937 / rho ** 5);
    tc = tc0 + tc1 + tc2;
    tc_ptrho = tc
    return tc_ptrho
```

## 5.3 Surface Tension

```python
def Surface_Tension_T(T):
    # IAPWS Release on Surface Tension of Ordinary Water Substance,
    # September 1994
    tc = 647.096; #K
    B = 0.2358;    #N/m
    bb = -0.625;
    my = 1.256;
    if T < 0.01 or T > tc: 
        Surface_Tension_T = np.nan; #"Out of valid region"
    tau = 1 - T / tc;
    Surface_Tension_T = B * tau ** my * (1 + bb * tau)
    return Surface_Tension_T
```
